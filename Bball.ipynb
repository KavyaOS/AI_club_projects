```python
import tensorflow as tf
import random
import math
```

```python
print(1+4)
```

```python
x = []
y = []
for i in range(100):
    height = random.random()
    stamina = random.random()
    
    stats = [height, stamina]
    x.append(stats)
    
    bball_score = height - stamina -1
    
    if bball_score > 0:
        y.append(1.0)
        
    else:
        y.append(0.0)
        
        
print(x,y)
```

```python
x = tf.Variable(x)
y = tf.Variable(y)

print(x)
```

```python
def sigmoid(x):
    return 1 / (1 + math.e**(-x))
```

```python
def call(x, w1, w2, b):
    return sigmoid(w1 * x[:,0] + w2 * x[:, 1] + b)
```

```python
def binary_crossentropy(y, y_hat):
    
    y_hat = 0.999*(y_hat + 0.0001)
    
    loss = -(y*tf.math.log(y_hat) + (1-y) * tf.math.log(1-y_hat))
    loss = tf.reduce_mean(loss)
    return loss
```

```python
w1 = tf.Variable(random.random())
w2 = tf.Variable(random.random())
b = tf.Variable(random.random())
```

```python
for i in range(1):
    with tf.GradientTape() as tape:
        y_hat = call(x, w1, w2, b)
        loss = binary_crossentropy(y, y_hat)
        
    gradient = tape.gradient(loss, [w1,w2,b])
    
    dw1 = gradient[0]
    dw2 = gradient[1]
    db = gradient[2]
    
    w1.assign(w1 - dw1 * 0.1)
    w2.assign(w2 - dw2 * 0.1)
    b.assign(b - db *0.1)
    
print(w1.numpy(), w2.numpy(), b.numpy())
```

```python
y_hat = call(x, w1, w2, b)

num_correct = 0
for y_i, y_i_hat in zip(y.numpy(), y_hat.numpy()):
    if abs((y_i - y_hat).any()) < 0.5:
            num_correct = num_correct + 1
            
print(num_correct)
```
