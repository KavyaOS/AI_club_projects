```python
from easyocr import Reader
```

```python
r = Reader(['en'])
```

```python
from PIL import Image, ImageDraw, ImageFont
```

```python
img = Image.open(path)
```

```python
img
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(img)
```

```python
draw = ImageDraw.Draw(img)
draw
```

```python
draw.rectangle(((88, 72), (158, 107)), fill=(0,0,0))
plt.imshow(img)
plt.show()
```

```python
font = ImageFont.truetype("arial.ttf", 30)
draw.text((88,72), "LIVE", (255,255,255), font=font)
plt.imshow(img)
plt.show()
```
